```
with_Emoticon=df[df["emoticon"]!="none"]
```

```
without_Emoticon=df[df["emoticon"]=="none"]
```

```
without_Emoticon
```

```
set_from_emo=with_Emoticon[0:200]
```

```
set_from_withoutEmo=without_Emoticon[0:200]
```

```
dataset=pd.concat([set_from_emo,set_from_withoutEmo],ignore_index=True)
```

```
dataset
```

```
def use_vader (line):
  anaylsis=sid.polarity_scores(line)
  Negitive=anaylsis['neg']*100
  Positive=anaylsis['pos']*100
  Neutral=anaylsis['neu']*100
  Values=[]
  Values.append(Negitive)
  Values.append(Positive)
  Values.append(Neutral)
  sentiment=""
  if anaylsis['compound'] >= 0.05 : 
    sentiment="Positive" 
  
  elif anaylsis['compound'] <= - 0.05 : 
    sentiment="Negative" 
  
  else : 
    sentiment="Neutral" 
  
  return (Values,sentiment)

```

```
arr,sen=use_vader(line)
```

```
arr
```

```
sen
```

```
vader_anaylsis=dataset
```

```
vader_anaylsis["Polarity"]=''
vader_anaylsis["Predicted_Sentiment"]=''
vader_anaylsis["Predict_VS_Rating"]=''
```

```
vader_anaylsis
```

```
vader_count_Negative=0
vader_count_Positive=0
vader_count_Neutral=0
for index,row in vader_anaylsis.iterrows():
  line=row['reviews.text']
  #print(line)
  pol,pre=use_vader(line)
  if pre=="Positive":
    vader_count_Positive=vader_count_Positive+1
    if row['reviews.rating']>=3:
      vader_anaylsis['Predict_VS_Rating'][index]='True'
    else:
      vader_anaylsis['Predict_VS_Rating'][index]='False'
  elif pre=="Negative":
    vader_count_Negative=vader_count_Negative+1
    if row['reviews.rating']<3:
      vader_anaylsis['Predict_VS_Rating'][index]='True'
    else:
      vader_anaylsis['Predict_VS_Rating'][index]='False'
  else:
    vader_count_Neutral=vader_count_Neutral+1
  vader_anaylsis['Polarity'][index]=pol
  vader_anaylsis['Predicted_Sentiment'][index]=pre
```

```
vader_anaylsis
```

```
true=vader_anaylsis[vader_anaylsis['Predict_VS_Rating']=='True']
false=vader_anaylsis[vader_anaylsis['Predict_VS_Rating']=='False']
```

```
print(f"True sentiments are {true.shape[0]}")
print(f"False sentiments are {false.shape[0]}")
```

```
print("Results")
print(f"Total labled positive:{vader_count_Positive}")
print(f"Total labled negative:{vader_count_Negative}")
print(f"Total labled neutral:{vader_count_Neutral}")
```

```
def use_TextBlob (line):
  anaylsis=TextBlob(line).sentiment
  Val=anaylsis[0]
  sentiment=""
  if anaylsis[0] > 0 : 
    sentiment="Positive" 
  
  elif anaylsis[0] < 0 : 
    sentiment="Negative" 
  
  else : 
    sentiment="Neutral" 
  
  return (Val,sentiment)

```

```
def use_Flair (line):
  anaylsis= flair.data.Sentence(line)
  flair_sentiment.predict(anaylsis)
  total_sentiment = anaylsis.labels
  Val=total_sentiment[0]
  string=str(total_sentiment[0])
  sentiment=""
  if "POSITIVE" in string: 
    sentiment="Positive" 
  
  elif "NEGATIVE" in string: 
    sentiment="Negative" 
  
  else : 
    sentiment="Neutral" 
  
  return (Val,sentiment)
```
